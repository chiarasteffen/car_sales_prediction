## 1. Setup & Laden des Datensatzes  
Import der benötigten Bibliotheken und Einlesen des bereinigten Datensatzes

```python
# 1.1 Imports
import pandas as pd
import numpy as np

# 1.2 Pfad zum bereinigten Datensatz
DATA_PATH = "../data/processed/cars_states_cleaned.csv"

# 1.3 Einlesen des Datensatzes
cars_states = pd.read_csv(
    DATA_PATH,
    parse_dates=["saledate"],
    infer_datetime_format=True
)

# 1.4 Grober Überblick
print("Shape des DataFrames:", cars_states.shape)
display(cars_states.head())
```

## 2. Kurzer Daten-Überblick  
Anzeige von `.info()`, Head und Basis-Statistiken zum Verifizieren nach Cleaning

```python
# 2.1 Struktur und Datentypen
cars_states.info()

# 2.2 Basis-Statistiken (numerisch)
display(cars_states.describe())

# 2.3 Basis-Statistiken (kategorisch)
display(cars_states.describe(include="object"))
```

## 3. Missing Values & Ausreisser (Kurz)  
Kurze Zusammenfassung der fehlenden Werte und Ausreisser in den finalen numerischen Features

```python
# 3.1 Fehlende Werte pro Spalte
missing_report = cars_states.isna().sum().to_frame("n_missing")
missing_report["pct_missing"] = 100 * missing_report["n_missing"] / len(cars_states)
print("Fehlende Werte (nur Spalten mit >0):")
display(missing_report[missing_report["n_missing"] > 0].sort_values("pct_missing", ascending=False))

# 3.2 Ausreißer-Detection für die numerischen Kern-Features
def outlier_bounds(series, k=1.5):
    q1, q3 = series.quantile([0.25, 0.75])
    iqr = q3 - q1
    return q1 - k * iqr, q3 + k * iqr

numeric_cols = ["sellingprice", "odometer", "condition", "mmr"]
outlier_summary = {}
for col in numeric_cols:
    lower, upper = outlier_bounds(cars_states[col].dropna())
    mask = (cars_states[col] < lower) | (cars_states[col] > upper)
    outlier_summary[col] = {
        "lower_bound": lower,
        "upper_bound": upper,
        "n_outliers": int(mask.sum()),
        "pct_outliers": float(mask.mean() * 100)
    }

print("Ausreißer-Zusammenfassung:")
for col, stats in outlier_summary.items():
    print(f"- {col}: {stats['n_outliers']} Ausreißer ({stats['pct_outliers']:.2f} %), Grenzen = [{stats['lower_bound']:.1f}, {stats['upper_bound']:.1f}]")
```

## 4. Feature Engineering  
Erstellung neuer Features für regionale Preisstatistiken, Saison, Text-Mining, Marktsegmente, Nutzungskennzahlen und Farb-Popularität

#### 4.1 Regionale Preisstatistiken  
Berechnung von `avg_price_state` (Durchschnittspreis) und `median_price_state` (Medianpreis) pro Bundesstaat und Merge ins Haupt-DataFrame

```python
# 4.1 Regionale Preisstatistiken

# 4.1.1 Preisstatistiken pro Bundesstaat berechnen
state_stats = (
    cars_states
    .groupby("StateName")["sellingprice"]
    .agg(avg_price_state="mean", median_price_state="median")
    .reset_index()
)

# 4.1.2 Merge der Statistik zurück in den Haupt-DataFrame
cars_states = cars_states.merge(
    state_stats,
    on="StateName",
    how="left"
)

# 4.1.3 Kontrolle: einige Beispiele anzeigen
print("Beispiele für regionale Preisstatistiken:")
display(state_stats.head())
```

```python
# Kontrolle 1: Direkter Vergleich für einen bekannten Staat, z. B. California
cal_fw = state_stats[state_stats["StateName"] == "California"].iloc[0]
sample = cars_states[cars_states["StateName"] == "California"][["StateName", "avg_price_state", "median_price_state"]].head(3)
print("Erwartete Werte für California:", cal_fw.to_dict())
display(sample)

# Kontrolle 2: Übersicht aller neuen Spalten im Haupt-DataFrame
display(cars_states[["StateName", "avg_price_state", "median_price_state"]].drop_duplicates().head(10))
```

```python
# Vorschau Datensatz mit neuen Spalten
print("Shape des erweiterten DataFrames:", cars_states.shape)
display(cars_states.head())
```

#### 4.2 Saison-Feature & Marken-Modell-spezifische Saisonpreise  
Einordnung von `sale_month` in Jahreszeiten und Berechnung des durchschnittlichen Verkaufspreises je `make`, `model` und `season`

```python
# 4.2 Saison-Feature & Marken-Modell-spezifische Saisonpreise

# Jahreszeit zuweisen (falls noch nicht geschehen)
def assign_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Fruehling"
    elif month in [6, 7, 8]:
        return "Sommer"
    else:
        return "Herbst"

cars_states["season"] = cars_states["sale_month"].astype(int).apply(assign_season)

# Durchschnittlicher Preis je Marke, Modell und Saison
make_model_season_price = (
    cars_states
    .groupby(["make", "model", "season"])["sellingprice"]
    .mean()
    .reset_index(name="mean_price_make_model_season")
)

# Kontrolle: Beispiel-Ausgabe
print("Beispiel: Durchschnittspreise für einige Make-Model-Season-Kombinationen")
display(make_model_season_price.head(10))

# Prüfung, wie viele Make-Model-Kombinationen in mehreren Seasons vorkommen
season_counts = (
    make_model_season_price
    .groupby(["make", "model"])["season"]
    .nunique()
    .reset_index(name="season_variants")
)
print("Anzahl Make-Model-Kombinationen, die in >1 Season verkauft wurden:",
      (season_counts["season_variants"] > 1).sum())
```

```python
# Prüfen, ob 'season' in den Spalten enthalten ist
print("season" in cars_states.columns)

# Ein paar Zeilen mit month und season anzeigen
display(cars_states[["sale_month", "season"]].head())

# Spaltenbezeichnung dataframe anzeigen
print("Spalten im DataFrame:", cars_states.columns.tolist())
```

##### 4.2.1 Erstellung und Export der saisonalen Mean-Price-Tabelle  
Erstellen eines separaten DataFrames mit `mean_price_make_model_season` für späteren Export und Merge

```python
# 4.2.1 Separaten DataFrame erstellen
mean_price_make_model_season = (
    cars_states
    .groupby(["make", "model", "season"])["sellingprice"]
    .mean()
    .reset_index(name="mean_price_make_model_season")
)

# Kontrolle: Head anzeigen
print("Beispiel für mean_price_make_model_season:")
display(mean_price_make_model_season.head())

# Export als CSV
mean_price_make_model_season.to_csv(
    "../data/processed/mean_price_make_model_season.csv",
    index=False
)
print("Exportiert nach ../data/processed/mean_price_make_model_season.csv")
```

##### 4.2.2 Merge der saisonalen Mean-Price-Werte ins Haupt-DataFrame  
Verknüpfen von `mean_price_make_model_season` über `make`, `model` und `season` mit `cars_states` und Zählen der fehlenden Werte

```python
# 4.2.2 Merge der saisonalen Mean-Price-Werte

# Merge auf make, model und season (links behalten)
cars_states = cars_states.merge(
    mean_price_make_model_season,
    on=["make", "model", "season"],
    how="left"
)

# Kontrolle: Anzahl der NaN-Werte in der neuen Spalte
n_missing_season_price = cars_states["mean_price_make_model_season"].isna().sum()
pct_missing_season_price = n_missing_season_price / len(cars_states) * 100
print(f"Fehlende mean_price_make_model_season: {n_missing_season_price} "
      f"({pct_missing_season_price:.2f} % der Zeilen)")
```

##### 4.2.3 Entfernen fehlender saisonaler Mean-Price-Werte  
Entfernen aller Zeilen, in denen `mean_price_make_model_season` fehlt

```python
# 4.2.3 Drop von Zeilen ohne mean_price_make_model_season
before_drop = len(cars_states)
cars_states = cars_states.dropna(subset=["mean_price_make_model_season"]).reset_index(drop=True)
after_drop = len(cars_states)

print(f"Entfernte Zeilen: {before_drop - after_drop}")
print(f"Verbleibende Zeilen: {after_drop}")
```

#### 4.3 Text-Mining-Indikatoren  
Erstellen von Binär-Flags (`has_sport`, `has_limited`, `has_lx` etc.) basierend auf Stichwörtern in den Spalten `trim` und `model`

```python
# 4.3 Text-Mining-Indikatoren

# Liste der Schlüsselwörter
keywords = ["Sport", "Limited", "LX", "SE", "Touring", "Premium"]

# Für jedes Keyword eine neue Binär-Spalte erzeugen
for kw in keywords:
    col_name = f"has_{kw.lower()}"
    # Prüfen in trim und model
    cars_states[col_name] = (
        cars_states["trim"].str.contains(kw, case=False, na=False) |
        cars_states["model"].str.contains(kw, case=False, na=False)
    ).astype(int)

# Kontrolle: Zeige die neuen Spalten für ein paar Zeilen
display(cars_states[["trim", "model"] + [f"has_{kw.lower()}" for kw in keywords]].head(10))
```

#### 4.4 Marktsegment-Buckets  
Top-10-`make` beibehalten und alle anderen in `"Other"` als `make_segment` zusammenfassen

```python
# 4.4 Marktsegment-Buckets

# 4.4.1 Top-10-Hersteller ermitteln
top10_makes = cars_states["make"].value_counts().nlargest(10).index.tolist()
print("Top-10-Hersteller:", top10_makes)

# 4.4.2 make_segment erstellen
cars_states["make_segment"] = cars_states["make"].apply(lambda x: x if x in top10_makes else "Other")

# 4.4.3 Kontrolle: Verteilung der Segmente
display(cars_states["make_segment"].value_counts(normalize=True).to_frame("anteil"))
```

#### 4.5 Nutzungskennzahlen & Farb-Popularität  
Berechnung von `miles_per_year` (Jahresdurchschnitt der gefahrenen Meilen) und `color_popularity` (Rang der Lackfarbe nach Häufigkeit)

```python
# 4.5 Nutzungskennzahlen & Farb-Popularität

# 4.5.1 miles_per_year berechnen
cars_states["vehicle_age"] = cars_states["sale_year"] - cars_states["year"]
# Vermeidung von Division durch Null, falls Baujahr == Verkaufsjahr
cars_states["vehicle_age"] = cars_states["vehicle_age"].replace(0, 1)
cars_states["miles_per_year"] = cars_states["odometer"] / cars_states["vehicle_age"]

# 4.5.2 color_popularity ermitteln
color_counts = cars_states["color"].value_counts()
# Erstelle ein Mapping von Farbe zu Rang (1 = häufigste)
color_rank = color_counts.rank(method="dense", ascending=False).astype(int)
# Mappe auf DataFrame
cars_states["color_popularity"] = cars_states["color"].map(color_rank)

# Kontrolle: ein paar Beispiele anzeigen
display(cars_states[["odometer", "vehicle_age", "miles_per_year", "color", "color_popularity"]].head(10))
```

## 5. Zusammenfassung & Export  
Erstellung der finalen Feature-Matrix und Export als CSV für das Modelltraining

```python
# 5.1 Auswahl der relevanten Spalten (falls gewünscht)
# Hier kannst Du alle Original-Features plus die neuen Features spezifizieren.
feature_cols = [
    # Numerisch
    "year", "condition", "odometer", "mmr", "sale_year", "sale_month", "sale_day", "sale_weekday",
    # Kategorial / Text
    "make_segment", "body", "transmission", "color", "interior",
    # Neue Features
    "avg_price_state", "median_price_state", "season",
    "mean_price_make_model_season",
    "has_sport", "has_limited", "has_lx", "has_se", "has_touring", "has_premium",
    "miles_per_year", "color_popularity"
]

# 5.2 Finale Feature-Matrix erstellen
cars_features_ready = cars_states[feature_cols].copy()

# 5.3 Export als CSV
output_path = "../data/processed/cars_features_ready.csv"
cars_features_ready.to_csv(output_path, index=False)
print(f"Feature-Matrix gespeichert unter: {output_path}")

# 5.4 Kurzer Blick auf das Ergebnis
print("Shape der finalen Feature-Matrix:", cars_features_ready.shape)
display(cars_features_ready.head())
```
