Explorative Datenanalyse und Feature-Generierung für Gebrauchtwagen-Preisvorhersage

## 1. Setup & Laden des Datensatzes  
Import der notwendigen Bibliotheken und Einlesen des bereinigten Datensatzes

```python
# 1.1 Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1.2 Pfad zum bereinigten Datensatz
DATA_PATH = "../data/processed/cars_states_cleaned.csv"

# 1.3 Einlesen des Datensatzes
cars_states = pd.read_csv(
    DATA_PATH,
    parse_dates=["saledate"],
    infer_datetime_format=True
)

# 1.4 Großer Überblick
print("Shape des DataFrames:", cars_states.shape)
display(cars_states.head())
```

## 2. Grober Daten-Überblick  
Ausgabe von Datentypen, Zusammenfassung und ersten statistischen Kennzahlen

```python
# 2.1 Datenstruktur und Datentypen
print("Datenstruktur und Datentypen:")
cars_states.info()

# 2.2 Deskriptive Statistik
print("\nDeskriptive Statistik (numerisch):")
display(cars_states.describe())

print("\nDeskriptive Statistik (alle Spalten):")
display(cars_states.describe(include="all"))
```

## 3. Datentypen optimieren  
Konvertieren der `sale_*`-Spalten in Ganzzahlen und Aktualisierung der numerischen Statistik ohne Datumsspalte

```python
# 3.1 sale_*-Spalten von float zu int konvertieren
for col in ["sale_year", "sale_month", "sale_day", "sale_weekday"]:
    cars_states[col] = cars_states[col].astype(int)
print("Datentypen nach Konvertierung:")
print(cars_states[["sale_year", "sale_month", "sale_day", "sale_weekday"]].dtypes)

# 3.2 Numerische Deskription ohne Datumsspalte
num_cols = ["year", "condition", "odometer", "mmr", "sellingprice",
            "sale_year", "sale_month", "sale_day", "sale_weekday"]
print("\nDeskriptive Statistik (numerisch, ohne saledate):")
display(cars_states[num_cols].describe())

# 3.3 Übersicht über den Zeitraum in saledate
print("\nZeitraum der Verkäufe:")
print("Start:", cars_states["saledate"].min())
print("Ende: ", cars_states["saledate"].max())
```
